```python
import requests
import bs4
```

```python
request=requests.get("https://www.flipkart.com/noise-loop-advanced-bluetooth-calling-1-85-display-550-nits-brightness-smartwatch/p/itmdfd159ab4efa7?pid=SMWGG6GFC9WUNEJK&lid=LSTSMWGG6GFC9WUNEJKGVGHVD&marketplace=FLIPKART&q=noise+smart+watch&store=ajy%2Fbuh&srno=s_1_24&otracker=search&otracker1=search&fm=organic&iid=0b9ef601-a558-412a-a304-17e2dbe20e9f.SMWGG6GFC9WUNEJK.SEARCH&ppt=None&ppn=None&ssid=yidge0gkcg0000001685354049205&qH=d519eaad7693519f")
```

```python
soup = bs4.BeautifulSoup(request.text)
soup
```

```python
reviews=soup.findAll('div',{'class':'ZmyHeo'})
all_review = []
for review in reviews:
    rr = review.get_text().replace("READ MORE" , "").strip()
    all_review.append(rr)
    print(rr+'\n')
```

```python
rating = soup.findAll('div' , {'class':"XQDdHH Ga3i8K"})
all_rating = []
for i in rating:
    rr = i.get_text()
    print(rr+'\n')
    all_rating.append(rr)
```

```python
review_title = soup.findAll('p',{'class':"z9E0IG"})
all_title=[]
for name in review_title:
    rr = name.get_text()
    all_title.append(rr)
    print(rr+'\n')
```

```python
names = soup.findAll('p' , {'class':'_2NsDsF AwS1CA'})
all_names = []
for name in names:
    rr = name.get_text()
    print(name.get_text()+'\n')
    all_names.append(rr)
```

```python
import pandas as pd
```

```python
data = {
    "reviews" : all_review,
    "title" : all_title,
    "names" : all_names,
    "rating" : all_rating
}
```

```python
df = pd.DataFrame(data)
```
